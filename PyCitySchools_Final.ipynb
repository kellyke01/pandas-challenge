# PyCity Schools Analysis

-Charter Schools Perform better than distrcit schools in terms of average math and average reading. The Charter Schools also outperformed district schools in the remaining categorites

-The smaller and medium sized schools performed better than larger schools in all areas. 

-Students at the schools with the largest budget per student were outperformed by the schools with smaller budgets.

```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("PyCitySchools1/schools_complete.csv")
student_data_to_load = Path("PyCitySchools1/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
# Calculate the percentage of students who passed reading (hint: look at how the math percentage was calculated)  
passing_reading_count = school_data_complete[(school_data_complete["reading_score"] >= 70)].count()["student_name"]
passing_reading_percentage = passing_reading_count / float(student_count) * 100
passing_reading_percentage
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passing_math_reading_count = school_data_complete[
    (school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(student_count) * 100
overall_passing_rate
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
district_summary = pd.DataFrame(
    { 
        "Total Schools": [school_count],
        "Total Students": [student_count],
        "Total Budget": [total_budget],
        "Average Math Score": [average_math_score],
        "Average Reading Score": [average_reading_score],
        "% Passing Math": [passing_math_percentage],
        "% Passing Reading": [passing_reading_percentage],
        "% Overall Passing": [overall_passing_rate]
    }
)


# Formatting
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)
district_summary["Average Math Score"] = district_summary["Average Math Score"].map("{:,.2f}".format)
district_summary["Average Reading Score"] = district_summary["Average Reading Score"].map("{:,.2f}".format)
district_summary["% Passing Math"] = district_summary["% Passing Math"].map("{:,.2f}".format)
district_summary["% Passing Reading"] = district_summary["% Passing Reading"].map("{:,.2f}".format)
district_summary["% Overall Passing"] = district_summary["% Overall Passing"].map("{:,.2f}".format)


# Display the DataFrame
district_summary
```

## Scores by School Spending

```python
per_school_summary.head()
```

```python
# max and min per student budget
print(per_school_summary["Per Student Budget"].max())
print(per_school_summary["Per Student Budget"].min())
```

```python
#Est. the bins
spending_bins = [0, 585, 630, 645, 680]
labels = ["<585", "585-630", "630-645", "645-680"]
```

```python
# Create a copy of the school summary since it has the "Per Student Budget" 
schoolSpendingDF = per_school_summary.copy()
schoolSpendingDF
```

```python
# Use `pd.cut` to categorize spending based on the bins.
schoolSpendingDF["Spending Ranges (Per Student)"] = pd.cut(per_school_capita, spending_bins, labels = labels)
schoolSpendingDF
```

```python
schoolSpendingDF["Average Math Scores"] = schoolSpendingDF["Average Math Scores"].astype(float)
schoolSpendingDF["Average Reading Scores"] = schoolSpendingDF["Average Reading Scores"].astype(float)
schoolSpendingDF["% Passing Math"] = schoolSpendingDF["% Passing Math"].str.replace("%","").astype(float)
schoolSpendingDF["% Passing Reading"] = schoolSpendingDF["% Passing Reading"].str.replace("%","").astype(float)
schoolSpendingDF["% Overall Passing"] = schoolSpendingDF["% Overall Passing"].str.replace("%","").astype(float)

schoolSpendingDF
```

```python
#first groupby spending ranges per student
spendingGroupedDF = schoolSpendingDF.groupby("Spending Ranges (Per Student)").mean("Per Student Budget")
spendingGroupedDF
```

```python
# grab the series
spendingMathSeries = spendingGroupedDF["Average Math Scores"]
spendingReadingSeries = spendingGroupedDF["Average Reading Scores"]
spendingPassMathSeries = spendingGroupedDF["% Passing Math"]
spendingPassReadingSeries = spendingGroupedDF["% Passing Reading"]
spendingPassSeries = spendingGroupedDF["% Overall Passing"]


                                                 
                                                 
```

```python
# Assemble into DataFrame
spendingSummaryDF = pd.DataFrame(
    {
        "Average Math Scores" : spendingMathSeries,
        "Average Reading Scores" : spendingReadingSeries,
        "% Passing Math" : spendingPassMathSeries,
        "% Passing Reading" : spendingPassReadingSeries,
        "% Overall Passing" : spendingPassSeries
    }
)

spendingSummaryDF["Average Math Scores"] = spendingSummaryDF["Average Math Scores"].map("{:.2f}".format)
spendingSummaryDF["Average Reading Scores"] = spendingSummaryDF["Average Reading Scores"].map("{:.2f}".format)
spendingSummaryDF["% Passing Math"] = spendingSummaryDF["% Passing Math"].map("{:.2f}".format)
spendingSummaryDF["% Passing Reading"] = spendingSummaryDF["% Passing Reading"].map("{:.2f}".format)
spendingSummaryDF["% Overall Passing"] = spendingSummaryDF["% Overall Passing"].map("{:.2f}".format)
        


# Display results
spendingSummaryDF
```

## Scores by School Size

```python
schoolSpendingDF2 = schoolSpendingDF.copy()
schoolSpendingDF2
```

```python
schoolSpendingDF2["Total Students"] = schoolSpendingDF["Total Students"].str.replace(",","").astype(int)
schoolSpendingDF2
```

```python
# Establish the bins.
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
# Categorize the spending based on the bins
# Use `pd.cut` on the "Total Students" column of the `per_school_summary` DataFrame.

schoolSpendingDF2["School Size"] = pd.cut(schoolSpendingDF2["Total Students"], size_bins, labels = labels)
schoolSpendingDF2
```

```python
sizeGroupedDF = schoolSpendingDF2.groupby("School Size").mean("Total Students")
sizeGroupedDF = sizeGroupedDF[["Average Math Scores", "Average Reading Scores", "% Passing Math", "% Passing Reading", "% Overall Passing"]]
sizeGroupedDF
```
